```python
import pandas as pd 
url = '/content/mtcars.csv'
data = pd.read_csv(url)
data.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation = data.corr()
print(correlation)
```

```python
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X=pd.DataFrame(data['wt']),y=data['mpg'])
print('Gia tri,Intercept la:',model.intercept_)
print('Gia tri coeffient la:',model.coef_)
model.score(X=pd.DataFrame(data['wt']),y=data['mpg'])
```

```python
data.plot(kind='scatter',x='wt',y='mpg',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['wt']))
plt.plot(data['wt'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X= pd.DataFrame([data['wt'],data['wt']**2]).T
model.fit(X,y=data['mpg'])
print('He so Intercept la:',model.intercept_)
print('He so coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(pd.DataFrame([data['wt'],data['wt']**2]).T,data['mpg']))
```

```python
import numpy as np

range = np.arange(1.5,5.5,0.1)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind = 'scatter', x = 'wt', y = 'mpg', figsize=(9,9),color='black',xlim=(0,7))
plt.plot(range, y_pred, color = 'blue')
```

```python
import pandas as pd
from sklearn import linear_model

data = pd.read_csv('/content/Advertising.csv')
model = linear_model.LinearRegression()

X = pd.DataFrame({'TV': data['TV'] ** 2,'radio': data['radio'] ** 2})

y = data['sales']

model.fit(X, y)
print('H s Intercept:', model.intercept_)
print('H s Coefficient:', model.coef_)
print(' chnh xc ca m hnh:', model.score(X, y))
```

```python
import numpy as np
range = np.arange(1.5, 300, 0.300)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='TV', y='sales', color='black', figsize=(9,9), xlim=(0, 300), ylim=(0, 30))
plt.plot(range, y_pred, color = 'blue')
```

```python
import numpy as np
range = np.arange(1.5, 300, 0.300)
X_range = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X_range)
data.plot(kind='scatter', x='radio', y='sales', color='black', figsize=(9,9), xlim=(0, 300), ylim=(0, 30))
plt.plot(range, y_pred, color = 'blue')
```
